```python
import numpy as np
import cv2 
```

# Image reading

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../Data/traffic_lights.jpg", 0)

plt.figure(figsize=(5, 10))
plt.imshow(img, cmap='gray')
plt.show()

print(img.shape)
```

# Canny filter

```python
edges = cv2.Canny(img, 20, 40)
high_thresh = cv2.Canny(img, 40, 110)

imgs = np.hstack((img, edges, high_thresh))

plt.figure(figsize=(15, 20))
plt.imshow(imgs, cmap='gray')
plt.show()
```
