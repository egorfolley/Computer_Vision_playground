```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

# Data uploading

```python
image = cv2.imread("../Data/coins.jpg", 0)

plt.imshow(image, cmap='gray')
plt.show()
```

# Adaptive thresholding

```python
ret,thresh_global = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
thresh_mean = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)

thresh_gaussian = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
```

# Display the results

```python
names = ['Original Image','Global Thresholding','Adaptive Mean Threshold','Adaptive Gaussian Thresholding']
images = [image,thresh_global,thresh_mean,thresh_gaussian]
```

```python
plt.figure(figsize=(10, 8))
plt.tight_layout()

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(names[i])
    plt.axis('off')
    
plt.show()
```
