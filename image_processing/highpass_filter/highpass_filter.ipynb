```python
import numpy as np
from scipy import ndimage
import cv2
```

# Reading image

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../data/Feynman_1.jpeg", 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(img_gray, cmap='gray')
```

# Kernels implementation

```python
kernel_3x3 = np.array([[-1, -1, -1],
                       [-1, 10, -1],
                       [-1, -1, -1]])

kernel_5x5 = np.array([[-1, -1, -1, -1, -1],
                       [-1, 1, 2, 1, -1],
                       [-1, 2, 8, 2, -1],
                       [-1, 1, 2, 1, -1],
                       [-1, -1, -1, -1, -1]])
```

```python
# Creating filters for edge detection

k3 = ndimage.convolve(img_gray, kernel_3x3)
k5 = ndimage.convolve(img_gray, kernel_5x5)
```

# Results

```python
blurred = cv2.GaussianBlur(img_gray, (11, 11), 0)
img_res = cv2.subtract(img_gray, blurred)
```

```python
plt.figure(figsize=(20, 20))

plt.subplot(1, 3, 1)
plt.imshow(k3, cmap='gray')

plt.subplot(1, 3, 2)
plt.imshow(k5, cmap='gray')

plt.subplot(1, 3, 3)
plt.imshow(img_res, cmap='gray')
```
