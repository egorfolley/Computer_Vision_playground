```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt 
```

# Import images

```python
img = cv2.imread("../Data/tony.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

# Averaging kernel for image smoothing

```python
avg_kernel = np.ones((3, 3), np.float32) / 9

filering_img = cv2.filter2D(img, -1, avg_kernel)
```

```python
plt.figure(figsize=(12, 12))

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(filering_img)
plt.axis('off')

plt.show()
```

# Gaussian kernel


```python
gaussian_kernel_x = cv2.getGaussianKernel(5,1) 
gaussian_kernel_y = cv2.getGaussianKernel(5,1) 
gaussian_kernel = gaussian_kernel_x * gaussian_kernel_y.T 

filtered_image = cv2.filter2D(img,-1,gaussian_kernel) 
```
