```python
import numpy as np
import cv2
```

# Image preparatiion

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../data/traffic_lights.jpg", 0)

plt.figure(figsize=(5, 10))
plt.imshow(img, cmap='gray')
```

# Bilateral filter

```python
img_filtered = cv2.bilateralFilter(img, 7, 50, 80)
```

### Applying the Canny filter

```python
edges = cv2.Canny(img, 60, 120)
edges_filtered = cv2.Canny(img_filtered, 40, 110)
```

```python
plt.figure(figsize=(15, 20))
plt.subplot(1, 3, 1).set_title("Image")
plt.imshow(img, cmap='gray')

plt.subplot(1, 3, 2).set_title("Canny filter")
plt.imshow(edges, cmap='gray')

plt.subplot(1, 3, 3).set_title("Canny filter with Bilateral filtering")
plt.imshow(edges_filtered, cmap='gray')

plt.show()
```
