```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

# Importing data

```python
img = cv2.imread('../data/traffic_lights.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

# Show separated colors

### Colors

```python
color_bounds = [
	([0, 10, 25], [255, 90, 45]),
	([40, 70, 20], [255, 255, 40]),
	([35, 160, 60], [65, 255, 130]),
    ([100, 210, 220], [110, 220, 240])
]

names_bounds = ['Red', 'Yellow', 'Green', 'Blue']
```

### Detection

```python
for (lower, upper), title in zip(color_bounds, names_bounds):
    # get lower and upper boundaries as arrays
    lower = np.array(lower, dtype=np.uint8)
    upper = np.array(upper, dtype=np.uint8)
    
    # find colors for specified boundaries and apply the mask
    mask = cv2.inRange(img, lower, upper)
    img_res = cv2.bitwise_and(img, img, mask=mask)
    
    imgs_stack = np.hstack([img, img_res])
    
    plt.figure(figsize=(10, 10))
    
    plt.title(title, fontsize=20)
    plt.imshow(imgs_stack)
    plt.show()
```
